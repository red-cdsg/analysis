## Plot
One little problem with Jupyter notebooks on VS Code is the theme of the plots. The theme of the plots does not go hand in hand with the theme of VS Code. To be precise, the light theme does not look good on a dark background. The data science documentation of VS Code shows how such a plot can be made beautiful. However, for E.D.A purposes, I will be using seaborn. This is better off on GitHub's light Jupyter theme.

```python
# A histogram is plot.
sns.set_theme(style="darkgrid")
sns.histplot(titanic3["age"], bins=20, kde=True)
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.legend(["Kernel Density Estimation", "Discrete Distribution"]);
```
